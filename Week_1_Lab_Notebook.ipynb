##Exercise 1
Recreate the FizzBuzz program - a classic task in coding interviews - in Python. FizzBuzz is a children’s game in which people in a circle count to 100 - with a twist. Your FizzBuzz program should loop from 1 to 50, and print, on a new line:

- “Fizz!” if the current number is divisible by 3
- “Buzz!” if the current number is divisible by 5
- “FizzBuzz!” if the current number is divisible by both 3 and 5 • Otherwise, print the current number.

```python
# Loop from 1 to 50
for i in range(1, 51):
    # Check if the number is divisible by both 3 and 5
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz!")
    # Check if the number is divisible by 3
    elif i % 3 == 0:
        print("Fizz!")
    # Check if the number is divisible by 5
    elif i % 5 == 0:
        print("Buzz!")
    # If the number is not divisible by either 3 or 5, print the number
    else:
        print(i)
```

##Exercise 2
This exercise will introduce you to the break and continue statements in Python.
- Create a new list with the numbers 3, 7, -9, 11, 2, -5, 0, 4.
- Create a variable, sum_pos, and set it equal to 0.
- Create a for loop that iterates through each of the numbers in the list. 
- If the current number is positive, add it to sum_pos.
- If the current number is negative, ignore it and move to the next number using the continue statement.
- If the current number is zero, stop the loop using break.


```python
# Create a new list with the numbers 3, 7, -9, 11, 2, -5, 0, 4.
lst = [3, 7, -9, 11, 2, -5, 0, 4]

# Create a variable, sum_pos, and set it equal to 0.
sum_pos = 0

# Create a for loop that iterates through each of the numbers in the list.
for num in lst:
    # If the current number is positive, add it to sum_pos.
    if num > 0:
        sum_pos += num
    # If the current number is negative, ignore it and move to the next number using the continue statement.
    elif num < 0:
        continue
    # If the current number is zero, stop the loop using break.
    elif num == 0:
        break

print(sum_pos)  
```

##Exercise 1
Build a function, `factorial(n)`, that takes in one input `n`, and returns `n!`.

```python
def factorial(n):
    # Base case: factorial of 0 is 1
    if n == 0:
        return 1
    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial(n-1)
```

##Exercise 2
Build a function, `fibonacci(n)`, that takes in one input `n`, and returns the `n`th Fibonacci number.

```python
def fibonacci(n):
    # The first Fibonacci number is 0
    if n == 0:
        return 0
    # The second Fibonacci number is 1
    elif n == 1:
        return 1
    # For all other n, the nth Fibonacci number is (n-1)th plus (n-2)th Fibonacci number
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
import pandas as pd
import numpy as np
titanic = pd.read_csv("train.csv")
titanic.head()
```

```python
#proportion of survivals
passengers_status = titanic['Survived']
print(f'The proportion of survivals is {round(np.sum(passengers_status)/len(passengers_status), 4)}')

#proportion of adolescent
adolescent_bool = titanic["Age"] < 18
adolescent_indicators =  adolescent_bool * 1
print(f'The proportion of adolescent is {round(np.sum(adolescent_indicators)/len(adolescent_indicators), 4)}')

#proportion of adolescent survivors
adolescent_status = titanic[adolescent_bool]['Survived']
print(f'The proportion of adolescent survivors is {round(np.sum(adolescent_status)/len(adolescent_status), 4)}')

#proportion of adult survivors
adult_status = titanic[[not state for state in adolescent_bool]]['Survived']
print(f'The proportion of adult survivors is {round(np.sum(adult_status)/len(adult_status), 4)}')
```
